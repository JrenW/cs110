First, I will create a Task object:

```python
import random

# from here, build a class of Task Object
class Task:
    # initialize a task of a given activity
    # if input activity_name = 'GLOBAL', this task is a global maltitasking task
    def __init__(self, actvty_name, actvty_length):  
        self.actvty = actvty_name
        self.actvty_length = actvty_length  # duration of entire activity, in minutes
        
        self.task_type = 'local' # default local
        self.multiTask = False  # default non-multitasking
        self.ID = '' # string
        self.description = '' # input string
        self.duration = 0 # task duration must be less than activity
        self.depends = []  # a list of dependent task IDs
        self.status = 'NOT_STARTED' 
            # three options: 'NOT_STARTED','IN_PROCESS', "COMPLETED"
 
        self.time_remain = 0  # remaining task time for multitasking tasks 
        self.sub_chunk = 0 # each multitask can be divided into sub chunks of k mins
        self.dynamic = False # a boolean for constant vs dynamic sub chunk duration
        
        self.pq_score = 0 # score of priority, initialized at zero
  
    # check if a task is multitasking type
    def is_multiTask(self):
        return self.multiTask


    # set task type to global multitasking one
    def is_global_task(self):
        if self.actvty == 'GLOBAL':
            self.task_type = 'global'
            self.multiTask = True    
            self.depends = []  # a global task has no dependencies, i.e. always ready
            return True
        else: 
            return False
    
        
    # specify basic task attributes
    def specify_task(self,lst_attrs):
        # input a list of task attributes in order of
        # [task_ID, description, duration, multiTask, dependencies, status]
        self.ID = lst_attrs[0]
        self.description = lst_attrs[1]
        self.duration = lst_attrs[2] # in minutes
        self.time_remain = lst_attrs[2] # same as total duration
        self.multiTask = lst_attrs[3]    
        if self.multiTask == False:
            self.sub_chunk = lst_attrs[2]  # only 1 subchunk for non-multitasking task
        self.depends = lst_attrs[4]  # a list of all dependent IDs
        self.status = lst_attrs[5]
        
    # parameterize multitasking tasks:
    def specify_multiTask(self, sub_chunk, is_dynamic):
        # if is_global_task, set respective attrs, if not (i.e. local multitask) nothing happens
        self.is_global_task()
        # add in sub_chunk 
        self.dynamic = is_dynamic # is_dynamic = True/False 
              # if True, each sub_chunk is a random integer from [1, sub_chunk]
        self.sub_chunk = sub_chunk # e.g. 2min to take photos out of total photo time of 60mins
              # for dynamic multitask, input sub_chunk value is the cap value
    
    
    def update_status(self):
        ### next, update task status depending on new self.time_remain 
        if 0 < self.time_remain < self.duration:
            if self.status == 'NOT_STARTED': 
                self.status = 'IN_PROCESS'
                #print(self.status)
        elif self.time_remain == 0:
            self.status = 'COMPLETED'
            #print(self.status)
        else: 
            #print('task alraedy completed')
            # reset time_remain to zero
            self.time_remain = 0
        
    
    # update remaining time & status of task depends on task type
    def update_task(self):  # minutes spent at current partition
        ### first, update time_remain
            # if non-multitasking
        if self.multiTask == False:
            self.time_remain -= self.time_remain  # spent all time at once
            
            # multitask
        else: 
            if self.dynamic == False:  # constant chunking
                self.time_remain -= self.sub_chunk
            else:    # dynamic chunking
                self.sub_chunk = max(1,random.randint(self.sub_chunk-3, self.sub_chunk+3))
                # update self
                if self.sub_chunk <= self.time_remain:
                    self.time_remain -= self.sub_chunk
                else: # if final chunk is reached, just subtract all remaining time
                    #print('Final chunk reached!')
                    self.time_remain -= self.time_remain
       
        # update status
        self.update_status()
    
    def update_nonMT(self, mins):   # manually update non-multitasking task by a given duratoin
        self.time_remain = self.duration
        rem = self.duration - mins
        self.time_remain -= rem
        #print(f'*****nonMT-time_remain: {self.time_remain}')
        self.sub_chunk = self.time_remain
        #print(f'*****nonMT-sub_chunk: {self.sub_chunk}')
        self.update_status()
    
    
    def reverse(self, sub_chunk):  # add back sub_chunk time, update time_remain and status 
        self.time_remain += sub_chunk
        # update status
        self.update_status()
    

    # check if task is ready: boolean return 
    def is_ready(self):
        if self.depends == []:
            return True
        else:
            return False

    # remove a dependency
    def remove_depends(self, task_ID):
        if task_ID in self.depends:
            self.depends.remove(task_ID)
            
    # change current pq score 
    def update_pqScore(self, new_pqScore):
        self.pq_score = new_pqScore
    
    # get length of dependencies 
    # this is global: any task in activity 2 should depend on completion of all tasks in activity 1
    def len_dpndncy(self):
        return len(self.depends)
    
    # change chunk length
    def change_chunk(self,chunk_length):
        self.sub_chunk = chunk_length
#         self.is_dynamic = is_dynamic
        
        
```

### Part C
#### [#ComputationalSolutions]
For multitasking tasks, how would you determine the optimal partitioning time?
- Test at least 2 different values or strategies and analyze their computational behavior.

#### My strategies: 
- my scheduler will run in an interpreter_style with 'short attention span' of 30 min;
- partitioning will be implemented via a priority queue structure (however for sake of shorter presentation, only loop is used here for an illustration);
- The scheduler will check across currently non-'COMPLETED' tasks, spot a task that is 'ready' (0 dependencies at current time stamp), calculate its priority queue score (according to some priority rules I made to reflect common sense of trip scheduling) and push it back into the priority queue;
- since I have declared 'global' tasks such as taking photos, those tasks will always remain in the priority queue by default (unless their total time is used up), the default pq = [GLB1,GLB2,GLB3,GLB4] of equal priority;
- so, all non-global tasks (i.e. local non-multitasking or local multitasking, *local = 'activity-specific'*) ,which are 'ready', effectively come into the priority queue with their pq score to 'challenge the status quo': 
        e.g1. a local non-multitasking task, JPT4 that has 0 dependencies would be given a high pq score, and once pushed in would surface at the top, so new pq = [JPT4,GLB1,GLB2,GLB3,GLB4], a non-multitasking task must be executed once pushed in; 
        e.g2. two local multitasking task, GGB3 and GGB4, come in with higher priority scores so current pq = [GGB3,GGB4,GLB1,GLB2,GLB3,GLB4]
- after we prepare the 30-min pq, the scheduler will 'execute' them in order of priority: 1st non-multitasking tasks; 2nd iterate between local multitasking tasks (e.g. GGB3 for 5min, GGB4 for 5min, GGB3 for 5min etc)
- once a task is 'COMPLETED' it is popped out of priority queue.
- if a local task remains in pq from last round, its pq will be updated to an even greater number at the end of 30-min window to make sure it is executed first in the next 30-min.



#### For this question, I will illustrate three types of partitioning (rigid to flexible) strategies:

    S1: all sub_chunks equal (5min interval) and static (no variation among chunks)
    S2: some sub_chunks have longer/shorter length, but all static
    S3: sub_chunks have varying length and are dynamic 
    
Let's say the scheduler has executed up to the point where DYG3-5 are all ready, i.e. one can multitask among DYG3(walk along art galleries), DYG4(listen to audio guide), DYG5(critique art pieces), plus all four global tasks that are always ready. 

According to my priority rule, the current pq = [DYG3, DYG4, DYG5, GLB1, GLB2, GLB3, GLB4], so the scheduler will first execute non-global ones in rotation ( 5min DYG3 --> 5min DYG4 --> 5min DYG5 --> 5min DYG3 etc) as the non-global ones will surface atop once pushed back into the pq

Let's examine the computational behavior of scheduler for three consecutive 30-min windows for each of the three strategies: 

```python
# prep work: prepare dummy duplicates of relevant task objects for simulation (status of actual objects not affected)

def dummy_tasks():
    dummy_lst = [['dmDYG3', 'walk along art galleries', 40, True, GGB+['DYG1','DYG2'], 'NOT_STARTED'],
                   ['dmDYG4', 'listen to audio guide', 20, True, GGB+['DYG1','DYG2'], 'NOT_STARTED'],
                   ['dmDYG5', 'critique an art piece with travel companions', 15, True, GGB+['DYG1','DYG2'], 'NOT_STARTED'],
                   ['dmGLB1', 'listen to songs from iphone', 60, True, [], 'NOT_STARTED'],
                   ['dmGLB2', 'facetime friends/family', 40, True, [], 'NOT_STARTED'],
                   ['dmGLB3', 'take photos from iphone', 60, True, [], 'NOT_STARTED'],
                   ['dmGLB4', 'chat with travel companions', 40, True, [], 'NOT_STARTED']]
    
    dmDYG3 = Task('visit de Young Museum', 150)  
    dmDYG4 = Task('visit de Young Museum', 150)   
    dmDYG5 = Task('visit de Young Museum', 150)
    dmGLB1 = Task('GLOBAL', 200)   
    dmGLB2 = Task('GLOBAL', 200)     
    dmGLB3 = Task('GLOBAL', 200)    
    dmGLB4 = Task('GLOBAL', 200)  
    
    dmTasks = [dmDYG3, dmDYG4, dmDYG5, dmGLB1, dmGLB2, dmGLB3, dmGLB4]

    
    for i in range(len(dmTasks)):
        dmTasks[i].specify_task(dummy_lst[i])
        if dmTasks[i].is_multiTask():  # if multitask, add in sub-chunk length
            dmTasks[i].specify_multiTask(sub_chunk=5, is_dynamic=False)
                # default choice of partition: each multitask is executed in 5-min constant interval
                
    return dmTasks


# next, a function to run dummy simulation according to priority rules
#*** here, priority rules are implemented through simple if statements, will apply to priority queue in section D
def run_simulation(dmTasks):
    for w in ['First', 'Second', 'Third']:
        print(f'{w} 30-min activity window: ')
        T_window = 30 
        while T_window >= 0:  
            for task in dmTasks[:3]:  # prioritize local multitasks first
                if task.time_remain > 0: 
                    task.update_task()

                elif task.time_remain == 0:
                    task = random.choice(dmTasks[3:])  # randomly pick a global task to do if a local task is finished
                    task.update_task()

                T_window -= task.sub_chunk
                if T_window < 0:
                    break 
                print(f'   {task.description} for {task.sub_chunk} mins    ({T_window}/30 mins remain)')

        print('  ')
```

**Strategy 3: different sub_chunks length equal and dynamic**

A Task object could have dynamic chunk length: given an input 'expected' sub_chunk value k, if dynamic==True, then thte object will randomly choose a number between [k-3, k+3]; in case k-3 is negative, the current chunk value will be assigned 1 (1 min minimal)

```python
### the max-heap class

def left(i):         # left(i): takes as input the array index of a parent node in the binary tree and 
    return 2*i + 1   #          returns the array index of its left child.

def right(i):        # right(i): takes as input the array index of a parent node in the binary tree and 
    return 2*i + 2   #           returns the array index of its right child.

def parent(i):       # parent(i): takes as input the array index of a node in the binary tree and
    return (i-1)//2  #            returns the array index of its parent


# Defining the Python class MaxHeapq to implement a max heap data structure.
# Every Object in this class has two attributes:
#           - heap : A Python list where key values in the max heap are stored
#           - heap_size: An integer counter of the number of keys present in the max heap
class maxHpq:
    """ 
    This class implements properties and methods that support a max priority queue data structure
    """  
    # Class initialization method. Use: heapq_var = MaxHeapq()
    def __init__(self):        
        self.heap       = []
        self.heap_size  = 0

    # This method returns the highest key in the priority queue. 
    #   Use: key_var = heapq_var.max()
    def maxk(self):              
        return self.heap[0]     
    
    # This method implements the INSERT key into a priority queue operation
    #   Use: heapq_var.heappush(key)
    def PQpush(self, key):   
        """
        Inserts the value of key onto the priority queue, maintaining the max heap invariant.
        """
        self.heap.append(-float("inf"))
        self.increase_key(self.heap_size,key)
        self.heap_size+=1
        

    # This method implements the INCREASE_KEY operation, which modifies the value of a key
    # in the max priority queue with a higher value. 
    #   Use heapq_var.increase_key(i, new_key)
    def increase_key(self, i, key): 
        if key < self.heap[i]:
            raise ValueError('new key is smaller than the current key')
        self.heap[i] = key
        while i > 0 and self.heap[parent(i)] < self.heap[i]:
            j = parent(i)
            holder = self.heap[j]
            self.heap[j] = self.heap[i]
            self.heap[i] = holder
            i = j    
            
    # This method implements the MAX_HEAPIFY operation for the max priority queue. The input is 
    # the array index of the root node of the subtree to be heapify.
    #   Use heapq_var.heapify(i)        
    def heapify(self, i):
        l = left(i)
        r = right(i)
        heap = self.heap
        if l <= (self.heap_size-1) and heap[l]>heap[i]:
            largest = l
        else:
            largest = i
        if r <= (self.heap_size-1) and heap[r] > heap[largest]:
            largest = r
        if largest != i:
            heap[i], heap[largest] = heap[largest], heap[i]
            self.heapify(largest)

    # This method implements the EXTRACT_MAX operation. It returns the largest key in 
    # the max priority queue and removes this key from the max priority queue.

    def PQpop(self):
        if self.heap_size < 1:
            raise ValueError('Heap underflow: There are no keys in the priority queue ')
        maxk = self.heap[0]
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self.heap_size-=1
        self.heapify(0)
        return maxk
    
    def reset(self):
        self.heap = []
        
    def show(self):
        return self.heap
        
```

```python
### the pqDict class
# create a dictionary object of {task.pq_score: task object}
class pqDict:
    def __init__(self):
        self.pqDict = {}     
        
    def add_item(self, key, value):
        self.pqDict.update({key : value})
        
    def delete_item(self, key):
        self.pqDict.pop(key, None)
        
    def update_key(self, old_score, new_score):
        taskObj = self.pqDict[old_score]
        taskObj.update_pqScore(new_score)
        self.pqDict.update({ taskObj.pq_score : taskObj })
        self.pqDict.pop(old_score, None)
    
    def getObj(self, pq_score):
        return self.pqDict[pq_score]
    
    def reset(self):
        self.pqDict = {}
        
    def show(self):
        return self.pqDict
```

```python
class Scheduler: 
    '''
    The scheduler only cares about global(always ready) vs local tasks (conditional)
    Scheduler priorities tasks with least number of currently dependent task IDs:
        all currently ready (0 denpendent) tasks are popped (removed) from input tasks;
            1) a scorer give a priority score for each 'ready' task object based on some heuristics;
            2) the pq_score of that object is pushed into a priority queue;
            3) add a { pq_score : taskObj } to pq dictionary;
        everytime, the maxScore is popped out of the priority queue, use the maxScore as key 
        to retrieve the 'primary task' from the dictionary;
        execute the task;
        if a task is COMPLETED after execution, 
            1) update dependencies in all remaining input tasks;
            2) remove key:value pair from pq dictionary;
            
        else, if a (multi)task is IN_PROCESS after sub_chunk execution,
            1) rescore the task pq_score, and push it back to current priority queue, 
            maintain maxHeap structure
        
        an activity window tracker is reset every 30min, 
            the Scheduler tries to maximize task assignment in every 30-min window
            by changing priority scores of global tasks when some 'few minutes left' warning is flagged
        
    '''
    def __init__(self): 
        self.global_tasks = []  # static
        self.local_tasks = []  # static
        self.current_tasks = []  # a dynamic bin to hold currently ready tasks
    
        self.pqHeap = maxHpq()  # maintain a max-heap of tasks' pq scores       
        self.sDict = pqDict() # a dictionary to retrieve a task object given its pq score
        
        self.wdw30 = 30 # a 30min clock,
        
        
    def read_in_tasks(self, Task_lst):  #  input a list of tasks (unordered)
        # classify tasks into global vs local 
        for task in Task_lst:
            if task.task_type == 'global':
                self.global_tasks.append(task) 
            else: 
                self.local_tasks.append(task)
                self.current_tasks.append(task)
                
                
    def scoreP(self, tskObj):  # heuristics to update priority score of a task based on task type                       
        
        if tskObj.is_global_task(): # least priority for global tasks
            tskObj.update_pqScore(-self.global_tasks.index(tskObj)) # negative of its index in global_task list
            
        else:  # else must be a local task
            lc = len(self.local_tasks)
            
            
            if tskObj.is_multiTask():                                 
                tskObj.update_pqScore( lc*1000 + self.local_tasks.index(tskObj))  
                                    # heuristics to make sure that multitasks stay in middle priority
            else:                   
                tskObj.update_pqScore( lc*1000000 + self.local_tasks.index(tskObj))  
                
                        ## later, for every update, current pq score is subtracted by 'lc'
        
    def initiate_pq_dict(self):
        # score and push all global tasks in
        for glb in self.global_tasks:
            self.scoreP(glb)  # update the task's pq_score
            new_score = glb.pq_score
            self.pqHeap.PQpush(new_score)  # push into max-heap
            self.sDict.add_item(glb.pq_score, glb)

            
   
    # pull ready only after a task is COMPLETED so dependencies gets updated for new readies
    def pull_ready(self):
        # first collect all currently ready tasks 
        rdy_lst = []
        for tsk in self.current_tasks:  # in currently unexecuted tasks           
            if tsk.is_ready():          
                rdy_lst.append(tsk)  # append to ready bin
    
             
        for tsk in rdy_lst:  
            self.scoreP(tsk)  # score the ready object         
            new_score = tsk.pq_score       
            self.pqHeap.PQpush(new_score)  # push new score into current max-heap
            self.sDict.add_item(new_score, tsk) # add a new key:value pair in dictionary
           
            for rm in self.current_tasks:   # detach all dependencies
                rm.remove_depends(tsk.ID)
            
            self.current_tasks.remove(tsk)  # pop out the item
            

        
        
    # push back a tskObj's pq_score after a task is still IN_PROCESS after current execution
    def push_back(self,tskObj):
        if tskObj.status == 'IN_PROCESS':
                # update its pq_score by subtracting by total length of local/global_list
            old_score = tskObj.pq_score  # get current score
                # calculate new score
            if tskObj.task_type == 'local': 
                new_score = old_score-len(self.local_tasks) 
            else:     # else a global task
                new_score = old_score-len(self.local_tasks) 
               
            # update task's pq_score
            tskObj.update_pqScore(new_score)        
            # next, update dictionary
            self.sDict.update_key(old_score, new_score)
            # finally, push the task's pq_score back to maxHeap
            self.pqHeap.PQpush(tskObj.pq_score)
            
            
    def show_ready(self):
        for k in self.current_tasks:
            if k.is_ready():
                print(k.ID)
   
    
    def only_globals(self):  #check if there is no local tasks in current pq
                        
        for pq in self.pqHeap.heap:  # (all global tasks are assigned a non-positive value)
            if pq > 0: 
                return False
        return True
        
    
    def one_window(self, wdw_number):  # for every 30min
        print('  ')
        print(f'******* ACTIVITY WINDOW {wdw_number}: 30mins *******')
        print('  ')
        
        self.wdw30 = 30
        
        while self.wdw30 > 0: 
            # pop out current max pq_score
            maxScore = self.pqHeap.PQpop()
            # retrieve the primary task given max pq_score
            prmy_task = self.sDict.getObj(maxScore)
            # execute the task (sub_chunk)
            prmy_task.update_task()
            #print(f'prmy_task.sub_chunk = {prmy_task.sub_chunk}')
            # update window timer
            self.wdw30 -= prmy_task.sub_chunk
            
            # check if window tracker still valid
            if self.wdw30 < 0: 
                
                prmy_task.reverse(prmy_task.sub_chunk) # undo execution 
                self.wdw30 += prmy_task.sub_chunk # undo window update
                         
                if prmy_task.is_multiTask() == False: 
                    # if a non-multitask, and total duration is more than windown time left
                    # manually update the portion, and push back same task
               
                    prmy_task.update_nonMT(self.wdw30)
                    # update window:
                    self.wdw30 = 0
                   
                    print(' ')
                    print(f'{self.wdw30} time wasted ')
                    # push back item
                    self.pqHeap.PQpush(prmy_task.pq_score)
           
                    return 
                    
                    
                
                else: # if a multitask subchunk is greater than window time left
                    # assign global tasks for last few minutes to minimize waste
                    tpls = self.special_handling()
                    for tpl in tpls:
                   
                        print(f'>>> {tpl[0].description} for {tpl[1]} mins, {self.wdw30}/30 mins remain')
                        print(' ')
                    # remeber to push back maxScore 
                    self.pqHeap.PQpush(prmy_task.pq_score)
       
                    print(f'{self.wdw30} mins wasted.')
                    print(' ')
                    break 
                
            if self.wdw30 == 0: 
                # remember to push back the popped maxScore
                self.pqHeap.PQpush(prmy_task.pq_score)
                #print(f'>>> {prmy_task.description} for {prmy_task.sub_chunk} mins, {self.wdw30}/30 mins remain')
                #print（' ')
                print(f'{self.wdw30} mins wasted.')
                print(' ')
                break
                
            # print task log
           
            print(f'>>> {prmy_task.description} for {prmy_task.sub_chunk} mins, {self.wdw30}/30 mins remain')
            print('  ')
            

            if prmy_task.status == 'COMPLETED':
                print(f'[-----Task {prmy_task.ID} is COMPLETED-----]')
                print('  ')
     
                if self.only_globals() == False :   
                        # hold on from pull_ready() if there are other local tasks in pq
                    continue
                    
                    
                else:  # else, there is only global tasks left (and there's still time)
                    # remove item from dictionary
                    self.sDict.delete_item(prmy_task.pq_score)
                    # pull new tasks that are ready in local_tasks
                    self.pull_ready()
        
                
            elif prmy_task.status == "IN_PROCESS":
                self.push_back(prmy_task)
                  
            else: 
                continue
             
    
    def special_handling(self):  # this is called when there are a few minutes left in 30-min window
        # get current time left in the activity window
        # randomly choose a global task to fill the gap
        spcls = []
        
        while self.wdw30 > 1:   # allow 1 min waste
            glb_task = random.choice(self.global_tasks)
            
                # if current subchunk is greater than remaining time, use up all window time
            if self.wdw30 < glb_task.sub_chunk: 
                glb_task.change_chunk(self.wdw30) # set sub_chunk value equal all remaining time
                glb_task.update_task()  # execute
                spcls.append((glb_task, self.wdw30))
                self.wdw30 = 0 
                break
                                                
            else: 
                self.wdw30 -= glb_task.sub_chunk
                spcls.append((glb_task, glb_task.sub_chunk))
        
        return spcls  # return a list of (taskObj, sub_chunk) tuples for task log reporting purpose
        
    
    def show_local(self):
        lcl_IDs = []
        for lcl in self.local_tasks:
            lcl_IDs.append(lcl.ID)  
        return lcl_IDs
    
    
    def show_global(self):
        glb_IDs = []
        for glb in self.global_tasks:
            glb_IDs.append(glb.ID)  
        return glb_IDs
    

    
```

**Computational Critique**

Pros: 
- priority queue is a superior data structure than simple array in a scheduler as it only cares about keeping the maxheap structure and make sure to return the largest value in queue (a scheduler is only interested in what's next) , whereas a loop would also take care of their ordering (i.e. the last item you append will always stay in the last slot); 
- by giving different task with a priority score, we can implement partitioning among multitasks without re-sorting items in a list every time a task is executed, instead, we simply push back a still IN_PROCESS multitask into the current pq, and it will automatically surface to its designated priority level (by maintaining the max heap structure;
- the scheduler results I got was quite human-like, with frequent attention shifts and juggling along 2-3 tasks at the same time (if they are all currently ready), some evidence:
![human like behavior](deYoung_excerpt.png)

- it is also surprising to see that, by implementing the pq structure itself results in some 'randomization' as compared to a sorted list due to its binary tree structure (e.g. on branch of high priority tasks, another branch of low priority tasks, the list won't appear sorted and same-level priority tasks may be executed in different order;
- here is some evidence of this dynamic features:
![evidence for dynamic task partitioning](JapTeaGarden_excerpt.png)


Cons:
- It would be better if we are allowed import of scipy.stats.norm so I could model the distribution of dynamic sub_chunk time by a Gaussian distribution which better reflects real-life behaivor;
- due to the dynamic feature and the unique way a priority queue is heapified, my priority score heuristics does not work every time. in my 20 tests of running run_Scheduler(  ), 2 times I get very weired behavior such as 'eat Japanese sesami mochi' for the next three 30mins window, but most of time, it returns a schedule that makes much human sense;
